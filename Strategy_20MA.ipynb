```python
from finlab.data import Data
# from talib import abstract

# 可以從 data 中讀取資料，拿前 100 日的
data = Data()
price = data.get('收盤價',200) 
volumn = data.get('成交金額',100)

volumn.iloc[-1]
```

```python
# 20 週線的平均
price_20MA = price[-100:].mean()
price_20MA
```

```python
# 林恩如策略 
# 前 5 天收盤價小於 20 週線
price_before5_20MA = price[-105:-5].mean()
strategy_before5_20MA = price.iloc[-5] < price_before5_20MA
 
# 今天收盤價大於 20 週線
price_today = data.get('收盤價',1)
price_20MA = price[-100:].mean()
strategy_20MA = price_today > price_20MA

# 今天開盤價大於 20 週線
price_open_today = data.get('開盤價',1)
price_open_20MA = price[-100:].mean()
strategy_open_20MA = price_today > price_open_20MA

# RSV 要大於 0.5
strategy_rsv = (price.iloc[-1] - price.iloc[-150:].min()) / (price.iloc[-150:].max() - price.iloc[-150:].min())

# 今天的成交是否帶量，超過5天內的平均的一倍
volumn_today = data.get('成交金額',1)
volumn_10DA = volumn[-5:].mean()
# volumn_std = volumn[-10:]
strategy_volumn = (volumn_today - volumn_10DA) > volumn_10DA

# 全部的策略
strategy_all =  strategy_before5_20MA & strategy_20MA & strategy_open_20MA & strategy_volumn 
# (volumn_today>10000000) &
# & (strategy_rsv > 0.5)
price_fill_strategy = price_today[strategy_all].dropna(1)

price_fill_strategy
for col in (price_fill_strategy.columns):
    print("https://goodinfo.tw/StockInfo/StockDetail.asp?STOCK_ID="+col)
 
```
